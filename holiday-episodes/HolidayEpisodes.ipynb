# Top-10 and bottom-10 holiday episodes

```python
df["ratings_diff"] = df["average_rating"] - df["parent_average_rating"]
```

```python
df_surp = df.sort_values(by="ratings_diff", ascending=False).head(5)[["original_title", "parent_original_title", 'year', 'season_number', 'episode_number', "average_rating", "parent_average_rating"]]
df_surp
```

```python
df_surp.apply(lambda x: f"{x.parent_original_title}: {x.original_title}, {str(int(x.year))}, S{str(int(x.season_number))}E{str(int(x.episode_number))}, {x.average_rating} vs {x.parent_average_rating}", axis=1)
```

```python
df_law = df.sort_values(by="average_rating", ascending=False).head(5)[["original_title", "parent_original_title", 'year', 'season_number', 'episode_number', "average_rating", "parent_average_rating"]]
```

```python
df_law.apply(lambda x: f"{x.parent_original_title}: {x.original_title}, {str(int(x.year))}, S{str(int(x.season_number))}E{str(int(x.episode_number))}, {x.average_rating}", axis=1)
```

```python
df_worst = df.sort_values(by="ratings_diff", ascending=True).head(10)[["original_title", "parent_original_title", 'year', 'season_number', 'episode_number', "average_rating", "parent_average_rating"]]
```

```python
df_worst.apply(lambda x: f"{x.parent_original_title}: {x.original_title}, {str(int(x.year))}, S{str(int(x.season_number))}E{str(int(x.episode_number))}, {x.average_rating} vs {x.parent_average_rating}", axis=1)
```
